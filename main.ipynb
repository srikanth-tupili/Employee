```python
import pathlib
import sqlite3
import pandas as pd


path = pathlib.Path().cwd() 


def emp_db(db_name, file_name, table_name):
    file_path = f"./{file_name}" 
    
    con = sqlite3.connect(db_name) 
    cur = con.cursor() 
    
    employee = pd.read_csv(file_path)
    employee.to_sql(table_name, con, index=False, if_exists='replace')
    
    results = cur.execute(f"SELECT * FROM {table_name}").fetchall()
    print(results)
    
    con.commit()
    con.close()
```
